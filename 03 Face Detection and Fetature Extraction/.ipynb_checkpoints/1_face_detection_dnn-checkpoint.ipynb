```python
import numpy as np
import cv2
```

```python
# load thte image
img = cv2.imread('./images/faces.jpg')

cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# detect faces using DNN

# load model

net = cv2.dnn.readNetFromCaffe('./models/deploy.prototxt.txt','./models/res10_300x300_ssd_iter_140000.caffemodel')
```

```python
# pass the image to the laoder

# extract blob

blob = cv2.dnn.blobFromImage(img,1,(300,300),(104,177,123),swapRB=False)
```

```python
blob.shape
```

```python
# set the blob as input

net.setInput(blob)

#run the model
detections = net.forward()
```

```python
detections.shape
```



(1,1,200,7)

- 1,1 = number of images

- 200 = faces detecte
- 7 values
    - 0: Image Number
    - 1 : binary (0 or 1) 0 means no face
    - 2 : Confidence score (0 to 1)
    - 3 : Start X
    - 4 : Start Y
    - 5 : End X
    - 6 : End Y

```python
h,w = img.shape[:2]

for i in range(0,detections.shape[2]):
    confidence = detections[0,0,i,2]
    if confidence > 0.5: # 0.5 == 50%
        # print(confidence)
        # bounding box (3:7)
        
        box = detections[0,0,i,3:7] # noramlised bounding box value means the values will be normalised to width and height of an image
        box = box*np.array([w,h,w,h]) # to find denormalisation
        box = box.astype(int) # convert to int because opencv doen't allow to plot
        # print(box)
        startx, starty, endx, endy = box
        
        # drawing the rectangle
        cv2.rectangle(img, (startx, starty), (endx,endy), (0,255,255))
        
        # use put text to put confidence score
        text = f'Face : {confidence*100:.2f}%'
        cv2.putText(img,text,(startx,starty-10),cv2.FONT_HERSHEY_PLAIN,1,(255,255,255))


cv2.imshow('Detected Imgae',img)
cv2.waitKey(0)
cv2.destroyAllWindows()        
```
